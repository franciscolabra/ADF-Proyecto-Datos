```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. Creamos un pequeño "mundo" de frases (nuestro corpus)
frases = [
    "El sistema de datos es muy lento",
    "El sistema de seguridad es rápido",
    "La seguridad de los datos es vital"
]

# 2. Configuramos el motor de ponderación (TF-IDF) --- prueba
# Usamos stop_words para ignorar palabras ultra-comunes si queremos,
# pero aquí lo dejaremos puro para ver cómo el algoritmo las pondera solo.
vectorizador = TfidfVectorizer()

# 3. Calculamos la ponderación
matriz_tfidf = vectorizador.fit_transform(frases)

# 4. Lo convertimos a una tabla para que sea fácil de leer para humanos
nombres_palabras = vectorizador.get_feature_names_out()
df = pd.DataFrame(matriz_tfidf.toarray(), columns=nombres_palabras)

print("Ponderación de palabras por cada frase:")
print(df)
```
